```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Dataframe

iris = sns.load_dataset('iris')
```

```python
# Basic Pair Plot
sns.pairplot(data=iris)
```

```python
# Change The Diagonal Pair Plot Kind
sns.pairplot(data=iris, diag_kind='kde')
```

```python
# Changing The Non Diagonal Plot Kind
sns.pairplot(data=iris, kind='reg')
```

```python
# Add Hue
sns.pairplot(data=iris, hue='species')
```

```python
# Create Pair Plot for Specific List of Variables
sns.pairplot(data=iris, x_vars=['petal_width', 'sepal_width', 'petal_length'], y_vars=['sepal_length', 'species'])
```

```python
# Show Only The Lower Triangle
sns.pairplot(data=iris, corner=True)
```

```python
# Styling Diagonal & Non Diagonal Plots

sns.pairplot(data=iris, diag_kws=dict(kde=True, color='blue', linewidth=1, edgecolor='cyan'), 
             plot_kws=dict(marker='d', linewidth=1, color='blue', edgecolor='cyan', s=65))

```

```python
# Create KDE on Top of Pair Plot
l = sns.pairplot(data=iris, plot_kws=dict(color='red'))
l.map_upper(sns.kdeplot)
l.map_lower(sns.kdeplot, fill=True)
```
