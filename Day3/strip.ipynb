```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(1)
```

```python
# Basic Strip Plot
sns.stripplot(data=df, y='Total')
```

```python
# Expand marker in Strip Plot

sns.stripplot(data=df, y='Total', jitter=0.4)
```

```python
# Draw line around the points
sns.stripplot(data=df, y='Total', jitter=0.4, linewidth=1, color='cyan')
```

```python
# Add Hue in Strip plot

plt.figure(figsize=(10, 8))
sns.stripplot(data=df, y='Total', hue='Product line',jitter=0.2, linewidth=1, palette='Set1')
```

```python
# Include 3 categorical varaible with Hue

plt.figure(figsize=(8, 5))
sns.stripplot(data=df, x='Payment', y='Total',hue='Gender',jitter=0.2, linewidth=1, palette='Set1')
```

```python

# Seperate each level of Hue
sns.stripplot(data=df, x='Payment', y='Total',hue='Gender',jitter=0.2, linewidth=1, palette='Set1', dodge=True)
```

```python

# Draw stripe on top of Violin plot
sns.stripplot(data=df , y='Total', x='Payment', jitter=0.1, palette='Set1')
sns.violinplot(data=df , y='Total', x='Payment', color='gray')
```

```python

# Draw stripe on top of Box plot
sns.stripplot(data=df , y='Total', x='Payment', jitter=0.1, palette='Set1')
sns.boxplot(data=df , y='Total', x='Payment', color='gray')
```
