```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(1)
```

```python
# Violin plot

sns.violinplot(y='Total', data=df)
```

```python
# Violin plot on two categorical and one numeric variable and use split

plt.figure(figsize=(13, 8))
sns.violinplot(x='Product line', y='Total', hue='Gender', data=df, palette='Set1', split=True)
plt.legend(bbox_to_anchor=(0.99, 1))
```

```python
# Box Horizontal Violin plot
sns.violinplot(x='Payment', y='Total', hue='Gender', data=df, palette='Set1', split=True, inner='quartile')
```

```python
# Draw line for each observation in violin plot

sns.violinplot(x='Payment', y='Total', hue='Gender', data=df, palette='Set1', split=True, inner='stick')
```

```python
# Change the amount of smoothing using bw attribute

sns.violinplot(x='Payment', y='Total', hue='Gender', data=df, palette='Set1', bw_method=0.5, split=True)
```

```python
# Cut out the extreme value

sns.violinplot(x='Payment', y='Total', hue='Gender', data=df, palette='Set1', bw_method=0.5, cut=0, split=True)
```
