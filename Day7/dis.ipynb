```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
```

```python
# Dis Plot w/ KDE

sns.displot(data=df, x='Total')
```

```python
# Dis Plot w/ KDE & rug

sns.displot(data=df, x='Total', rug=True, kde=True)
```

```python
# Styling KDE & Rug

sns.displot(data=df, x='Total', rug=True, kde=True, rug_kws={'height':0.05})
```

```python
# Add Hue
sns.displot(x='gross income', hue='Product line', data=df, multiple='stack', kde=True)
```

```python
# Create Seperate Plot 1

sns.displot(x='gross income', hue='Product line', data=df, multiple='stack', kde=True, col='Gender')
```

```python
# Create Seperate Plot 2
sns.displot(x='gross income', hue='Product line', data=df, multiple='stack', kde=True, col='Gender', row='Payment')
```

```python
# Create ECDF w/ Dis Plot 
sns.displot(x='Total', hue='Payment', data=df, kind='ecdf', stat='count')
```
