```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Dataframe
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Basic Joint Plot
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
# Custom Kind
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='resid')
```

```python
# Grouping Based on a Categorical Variable
sns.jointplot(x='petal_length', y='petal_width', hue='species', data=iris)
```

```python
# Formatting Individual Plot

sns.jointplot(x='petal_length', y='petal_width', data=iris, 
              joint_kws=dict(marker='D',
                             color='cyan',
                             linewidth=1,
                             edgecolor='blue'),
              marginal_kws=dict(color='violet',
                                linewidth=1,
                                edgecolor='magenta',
                                kde=True))
```

```python
# Formatting All Plot

sns.jointplot(x='petal_length', y='petal_width', hue='species', 
              data=iris, palette='Set1')
```

```python
# Styling Joint Plot
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, 
              height=5,
              ratio=2,
              marginal_ticks=True)
```

```python
# KDE & Rug on Top of Joint Plot

l = sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
l.plot_joint(sns.kdeplot, fill=True)
l.plot_joint(sns.rugplot, height=0.08, color='blue')
```
