```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Datafrane
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx")
df.columns = df.columns.str.lower()
df.head()
```

```python
df.outlet_location_type.unique()
```

```python
df.outlet_size.unique()
```

```python
# Basic Facetgrid
lernerea = sns.FacetGrid(df, col='outlet_size')
lernerea.map_dataframe(sns.lineplot, x='outlet_year', y='sales')
```

```python
# Change The Plot Type
# plt.figure(figsixze=(15, 8))
lernerea = sns.FacetGrid(df, col='outlet_size')
lernerea.map_dataframe(sns.barplot, x='outlet_location_type', y='sales', palette='Set1')
```

```python
# Styling Facetgrid
lernerea = sns.FacetGrid(df, col='outlet_size')
lernerea.map_dataframe(sns.lineplot, x='outlet_year', y='sales', 
                       errorbar=None, linewidth= 1.5, linestyle='--',
                       color='blue')
lernerea.set_axis_labels('Year', 'Sales')
lernerea.set_titles(col_template='{col_name} Size')
```
