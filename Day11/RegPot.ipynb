```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataframe
tips = sns.load_dataset('tips')
tips.head(1)
```

```python
# Basic Reg Plot
sns.regplot(data=tips, x='total_bill', y='tip')
```

```python
# Styling Reg Plot
sns.regplot(data=tips, x='total_bill', y='tip', marker='d',
            line_kws=dict(color='blue', linestyle='--', linewidth= 1.5),
            scatter_kws=dict(color='cyan', edgecolor='blue', s=50))
```
