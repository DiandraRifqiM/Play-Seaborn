```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Datframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx").sample(n=50)
df.columns = df.columns.str.lower()
df.head(1)
```

```python
# Basic Rel Plot
sns.relplot(data=df, x='outlet_year', y='sales')
```

```python

# Custom Kind in Rel Plot
sns.relplot(data=df, x='outlet_year', y='sales', kind='line', errorbar=None)
```

```python
# Categorical Rel Plot
sns.relplot(data=df, x='outlet_year', y='sales', hue='outlet_size', 
            palette=['cyan', 'lime', 'blue'], edgecolor='black', linewidth=0.5)
```

```python

# Seperate Rel Plot Columnsn
sns.relplot(data=df, x='outlet_year', y='sales', s=100, hue='outlet_size',
            col='tier', palette='Set1')
```

```python

# Wrap Column in Rel Plot
sns.relplot(data=df, x='outlet_year', y='sales', s=100, hue='outlet_size',
            col='tier', palette='Set1', row='outlet_size')
```
