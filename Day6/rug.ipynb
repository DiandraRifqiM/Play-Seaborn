```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
```

```python
# Basic Rug Plot

sns.rugplot(x='gross income', data=df)
```

```python
# Two Variables Rug Plot
sns.rugplot(x='Total', y='gross income', data=df)
```

```python
# Categorical Rug Plot
sns.rugplot(x='Total', y='gross income', hue='Product line', data=df, palette='Set1')
```

```python
# Change The Height Rug
sns.rugplot(x='Total', y='gross income', hue='Product line', data=df, palette='Set1', height=0.08)
```

```python
# Combine w/ KDE 
sns.rugplot(x='Unit price', y='gross income', hue='Gender', data=df, palette='Set1', height=0.05)
sns.kdeplot(x='Unit price', y='gross income', hue='Gender', data=df, fill=True, thresh=0.2)
```

```python
# Combine w/ Scatter
plt.figure(figsize=(13, 8))
sns.rugplot(x='Unit price', y='gross income', hue='Gender', data=df, height=0.05)
sns.scatterplot(x='Unit price', y='gross income', hue='Gender', data=df)
```
