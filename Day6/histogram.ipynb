```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(1)
```

```python
# Basic Histogram

sns.histplot(x='Total', data=df, color='skyblue', edgecolor='black', linewidth=0.5)
```

```python
# Change Bins width in Histogram
sns.histplot(x='Total', data=df, color='cyan', binwidth=50, linewidth=0.5)
```

```python
# Change Number of Bins & Interval
plt.figure(figsize=(15, 5))
sns.histplot(x='Total', data=df, color='skyblue', bins=np.arange(0, 1100, 50))
plt.xticks(np.arange(0, 1100, 50))
```

```python
# Combine w/ KDE

plt.figure(figsize=(15, 8))
sns.histplot(x='Product line', data=df, color='blue', linewidth= 1, kde=True)
```

```python
# Use a Categorical Variable in Hue & Stack it With Multiple Argument

plt.figure(figsize=(15, 6))
sns.histplot(x='Product line', hue='Gender', data=df, multiple='stack', palette='Set2')
```

```python
# Make a Step/Poly Using Element Arguement & Change The Fill
sns.histplot(x='Total', hue='Payment', data=df, element='poly')
```

```python
# Use Categorical Variable & Shrink it
plt.figure(figsize=(10, 8))
sns.histplot(x='Payment', hue='Gender', data=df, shrink=0.5, multiple='stack')
```

```python
# Create a Revariate Histogram
plt.figure(figsize=(15, 8))
sns.histplot(x='Total', y='gross income', data=df, color='magenta')
```
