```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts

pd.set_option('display.max_rows', None)
```

```python
# Dataframe
df=pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df1 = df[['Unit price', 'Quantity', 'Total', 'gross income']]
```

```python
# Basic KDE
sns.kdeplot(x='Total', data=df)
```

```python
# Create KDE For All Numeric Variables in DF
sns.kdeplot(data=df1)
```

```python
# Adjust The KDE Smoothing
sns.kdeplot(x='Total', data=df, bw_adjust=0.5)
```

```python
# Group The KDE on a Category Variables
sns.kdeplot(x='Total', hue='Product line', data=df)
```

```python
# Stack KDE on a Category
sns.kdeplot(x='Total', hue='Product line', data=df, multiple='stack')
```

```python
# Use Log Scalling to Map The Variable in KDE
sns.kdeplot(x='Total', data=df, log_scale=True)
```

```python
# Styling KDE
sns.kdeplot(x='Total', hue='Product line', data=df, multiple='stack', 
            palette='Set1', linewidth=0.8, alpha=True)
```

```python
# Create Bivariate KDE

sns.kdeplot(x='Unit price', y='gross income', data=df, hue='Gender', fill=True, thresh=0.2)
```
