```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx")
df.columns = df.columns.str.lower()
# df[df['outlet_year'] == 1988]
```

```python
# Pivot
mart = df.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
mart
```

```python
# Basic Heat Map
sns.set_style('white')
sns.heatmap(mart, square=True)
```

```python
# Showing number data
sns.set_style('white')
sns.heatmap(mart, annot=True, fmt='.0f')
```

```python
# Styling annot
sns.set_style('white')
sns.heatmap(mart, annot=True, fmt='.0f',
            annot_kws=dict(size=15, weight='bold'),
            linecolor='black', linewidths=0.5)
```

```python
sns.heatmap(mart.corr(), vmin=1, vmax=1)
```
