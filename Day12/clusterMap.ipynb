```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Datframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx")
df.columns = df.columns.str.lower()
dfPiv = round(df.pivot_table(index = 'outlet_year', columns= 'tier', values= 'sales'), 2)
dfPiv.fillna(0, inplace=True)
dfPiv.head()
```

```python
# Creating Basic ClusterMap
sns.clustermap(dfPiv)
```

```python
# Row Only Cluster
sns.clustermap(dfPiv, col_cluster=False)
```

```python
# Column Ony Cluster
sns.clustermap(dfPiv, row_cluster=False)
```

```python
# Styling Custer
sns.clustermap(dfPiv, annot=True,
               fmt='.0f')
0
```

```python
# Change Z Score

sns.clustermap(dfPiv, z_score=0, annot=True, fmt='.0f')
```

```python
# Scaling Cluster
sns.clustermap(dfPiv, annot=True,
               fmt='.0f', standard_scale=0.5)
```

```python
# Styling Line Cluster
sns.clustermap(dfPiv, annot=True,
               fmt='.0f', standard_scale=0.5,
               linewidth=0.5,)
```
