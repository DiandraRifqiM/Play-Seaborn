```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(1)
```

```python
# Basic Cat Plot
sns.catplot(x='Payment', y='Total', data=df, palette='Set1')
```

```python
# Cat Plot w/ Other Type
sns.catplot(x='Payment', y='Total', hue='Customer type', data=df, kind='bar', palette='Set1')
```

```python
# Seperate Cat Plot
sns.catplot(x='Payment', y='Total', col='Customer type', data=df, palette='Set1', 
            kind='swarm', edgecolor='cyan', linewidth=0.5, size=6)
```
