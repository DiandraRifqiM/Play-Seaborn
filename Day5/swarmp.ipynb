```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
```

```python
# Basic Swarm Plot
sns.swarmplot(y='Total', data=df)
```

```python
# Create Swarm Plot Group by Categorical

sns.swarmplot(y='Total', x='Payment', data=df, palette='Set1')
```

```python
# Seperate Hue on Swarm Plot

sns.swarmplot(data=df, y='Total', x='Payment', hue='Gender', dodge=True)
```

```python
# Styling Swarm Plot

plt.figure(figsize=(8, 8))
sns.swarmplot(data=df, y='Total', x='Payment', hue='Gender', dodge=True, 
              marker='D', linewidth=0.5, palette='Set1', edgecolor='cyan')
```

```python
# Overlay Swarm PLot (Violin)

sns.boxplot(y='Total', x='Payment', data=df, palette='Set1')
sns.swarmplot(y='Total', x='Payment', data=df, color='gray')
```

```python
# Overlay Swarm PLot (Box)

sns.violinplot(y='Total', x='Payment', data=df, palette='Set1')
sns.swarmplot(y='Total', x='Payment', data=df, color='white')
```
