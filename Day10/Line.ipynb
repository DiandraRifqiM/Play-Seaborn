```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx")
df.head(1)
```

```python
# Basic Line Plot & Try Different Estimator
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Outlet_Year', y='Sales', errorbar=None, estimator='mean')
```

```python
# Test w/ Different Confidence Intervals & Different Number of Bootstrapping
sns.lineplot(data=df, x='Outlet_Year', y='Sales', hue='Outlet_Size', errorbar=None, n_boot='sum')
```

```python
# Grouping
sns.lineplot(data=df, x='Outlet_Year', y='Sales', hue='Outlet_Size', 
             errorbar=None, estimator='sum', palette=['cyan','blue','lime'])
```

```python
# Grouping & Styling
sns.lineplot(data=df, x='Outlet_Year', y='Sales', hue='Outlet_Size',
             style='Outlet_Size', markers=True, dashes=False)
```

```python
# Grouping & Sizing
plt.figure(figsize=(10, 7))
sns.lineplot(data=df, x='Outlet_Year', y='Sales', hue='Outlet_Size',
             style='Outlet_Size', size='Tier', sizes=(.3,3))
```

```python
# Use Different Sizing Parameters On Same || Different Variables
plt.figure(figsize=(10, 7))
sns.lineplot(data=df, x='Outlet_Year', y='Sales', hue='Tier',
             size='Outlet_Size', ci=None, markers=True, 
             palette='Set1')
```
