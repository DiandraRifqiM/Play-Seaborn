```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Datframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\mart_train_sample.xlsx").sample(50)
df.head(1)
```

```python
# Basic Scatter Plot
sns.scatterplot(data=df, x='Item_MRP', y='Sales')
```

```python
# Categorical Scatter Plot
sns.scatterplot(data=df, x='Item_MRP', y='Sales', hue='Outlet_Size')
```

```python
# Categorical Scatter Plot & Styling 1
sns.scatterplot(data=df, x='Item_MRP', y='Sales', hue='Outlet_Size',
                style='Outlet_Size',
                markers={"High":"D", "Medium":"^", "Small":"v"})
```

```python
# Categorical Scatter Plot & Styling 2
sns.scatterplot(data=df, x='Item_MRP', y='Sales',
                hue='Outlet_Size', size='Tier', palette='Set1', 
                linewidth=0.5, edgecolor='black')
```

```python
# Numeric Scatter 
sns.scatterplot(data=df, x='Item_MRP', y='Sales', 
                hue='Outlet_Size', size='Tier',
                palette='Set2', markers={"High":"D", "Medium":"^", "Small":"v"})
```
