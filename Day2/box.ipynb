```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(2)
```

```python
# Box Plot w/ One Numeric Variable
sns.boxplot(x='Total', data=df, palette='Set1')
```

```python
# Box Plot w/ One Numeric Variable by Categorical
sns.boxplot(x='Payment', y='Total', data=df, palette='Set1')
```

```python
# Box Plot w/ One Numeric Variable by 2 Categorical
sns.boxplot(x='Payment', y='Total', hue='Gender', data=df, palette='Set1', saturation=0.8)
```

```python
# Box Plot Mean Marker w/ Showmeans and Styling w/ Meanprops
sns.boxplot(x="Payment", y='Total', hue='Gender', data=df, palette='Set1', saturation=0.8, showmeans=True, 
            meanprops={'marker':'o',
                       'markerfacecolor':'lime',
                       'markersize':8,
                       'markeredgecolor':'black'
                       })
```

```python
# Horizontal Box Plot
sns.boxplot(y='Payment', x='Total', hue='Gender', data=df, palette='Set1', saturation=0.8, showmeans=True,
            meanprops={'marker':'D',
                       'markersize':8,
                       'markerfacecolor':'lime',
                       'markeredgecolor':'black'})
```

```python
# Boxplot Styling
plt.figure(figsize=(13, 8))
sns.boxplot(x='Product line', y='Total', data=df, palette='Set1', saturation=0.8, linewidth=1.5, showmeans=True,
            meanprops={'marker':'o',
                       'markersize':8,
                       'markerfacecolor':'lime',
                       'markeredgecolor':'black'})
```
