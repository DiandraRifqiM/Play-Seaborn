```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Display seabord data sets
sns.get_dataset_names()
titanicData = sns.load_dataset('titanic')
titanicData.head()
```

```python
# Supermarket Dataframe
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df.head(2)
```

```python
# Bar Plot
plt.figure(figsize=(13, 8))
sns.barplot(x='Product line', y='Total', data=df, palette='Set1')
```

```python
# Add Hue Bar Plot
plt.figure(figsize=(13,8))
sns.barplot(x='Product line', y='Total', hue='Gender', data=df, palette='Set1', hue_order=['Male','Female'])
```

```python
# Horizontal Bar Plot
plt.figure(figsize=(10, 8))
sns.barplot(x='Total', y='Product line', data=df, palette='Set1')
```

```python
# Add Cap on Error Bar
plt.figure(figsize=(13, 8))
sns.barplot(x='Product line', y='Total', data=df, palette='Set1', capsize=0.2)
```

```python
# Remove Error Bar
plt.figure(figsize=(13, 8))
sns.barplot(x='Product line', y='Total', data=df, palette='Set1', ci=None)
```

```python
# Add Saturation
plt.figure(figsize=(13, 8))
sns.barplot(x='Product line', y='Total', data=df, palette='Set1', ci=None, saturation=1)
```

```python
# Add Estimator
plt.figure(figsize=(13, 8))
sns.barplot(x='Product line', y='Total', data=df, palette='Set1', ci=None, saturation=1, estimator=np.median)
```
