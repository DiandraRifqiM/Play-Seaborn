```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# DataFrame
df = pd.read_csv(r"D:\Python Course\Playing Seaborn\xlsx\retail_data.csv")
df.head()
```

```python

# Fill missing Age with the mean age.
meAge = int(round(df["Age"].mean()))
df["Age"] = df["Age"].fillna(meAge)
df["Age"] = round(df["Age"], 0)
df.head()
```

```python
# Fill missing Product/Price logically (e.g., with mode or median).
modProd = df["Product"].mode()
mePrice = round(df["Price"].mean(), 1)
df["Product"] = df["Product"].fillna("Laptop")
df["Price"] = df["Price"].fillna(mePrice)
df.head()
```

```python
# Change gender values to F/M

df["Gender"] = df["Gender"].replace("Male", "M")
df["Gender"] = df["Gender"].replace("female", "F")
df["Gender"] = df["Gender"].str.capitalize()
df.head()
```

```python
# Fix typos in Category
df["Category"] = df["Category"].replace("electronic", "Electronics")
df["Category"] = df["Category"].str.capitalize()
df.head()
```

```python
# Standardize Payment methods typing
df["Payment"] = df["Payment"].str.capitalize()
df.head()
```

```python
# Fix date time
df["OrderDate"] = df["OrderDate"].replace("01/16/2023", "2023-01-16")
df["OrderDate"] = df["OrderDate"].str.replace("/", "-")
df["OrderDate"] = pd.to_datetime(df["OrderDate"], errors="coerce")
df.head()
```

```python
# Drop entire row where date == null
df = df.dropna(subset=["OrderDate"])
df[df["OrderDate"].isna()]
```

```python
# Drop entire row where quantity < 0
df.loc[df["Quantity"] < 0, "Quantity"] = np.nan
df = df.dropna(subset=["Quantity"])
df[df["Quantity"].isna()]
```
