```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")


# DataFrame
df = pd.read_excel(r"D:\Python Course\Playing Seaborn\xlsx\supermarket_sales.xlsx")
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
# Create Countplot
sns.countplot(x='Gender', data= df, palette= 'Set1')
```

```python
# Product Category Countplot

plt.figure(figsize=(13, 8))
sns.countplot(x = 'Product line', data= df, palette='Set1')
```

```python
# Horizontal Bar Plot

sns.countplot(y = 'Payment', data = df, palette='husl')
```

```python
# Hue
plt.figure(figsize=(8, 7))
sns.countplot(x = 'Payment', hue='Gender', data= df, palette='Set1')
```
